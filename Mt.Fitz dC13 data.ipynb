```python
#first step: import the tools you will be using
import pandas as pd
from matplotlib import pyplot as plt

#import the plot style
#search plot styles: print(plt.style.available)
plt.style.use('seaborn-darkgrid')

#import the data (panda reads it like a table)
data = pd.read_csv('dC13 data.csv')

#define the data variables
d13C = data['d13C_carb']
sample_age = data['sample_age']

#plot the data on a scatter plot
fig1=plt.scatter(d13C, sample_age, edgecolor='black', linewidth= 1, alpha = 0.5)

#connect the dots
plt.plot(d13C, sample_age)

#label the x and y axis
plt.xlabel('d13C_carb')

plt.ylabel('Sample Age (Yrs)')

#set the size of the x and y axes
def set_size(w,h, ax=None):
    """ w, h: width, height in inches """
    if not ax: ax=plt.gca()
    l = ax.figure.subplotpars.left
    r = ax.figure.subplotpars.right
    t = ax.figure.subplotpars.top
    b = ax.figure.subplotpars.bottom
    figw = float(w)/(r-l)
    figh = float(h)/(t-b)
    ax.figure.set_size_inches(figw, figh)
#this is where you actually input the size ratio idk what the "if not" madness above does but it doesnt work without it
set_size(5,10)

#invert the y-axis so that the strat height makes more sense 
#first set the axis limits... note: gca = 'get current axis' and gcf = "get current figure"
#then plot the inverted y-axis
fig1 = plt.gca()
#plt.gca().invert_yaxis()

#note: delta in python: '$\delta^{18}$O 
```
